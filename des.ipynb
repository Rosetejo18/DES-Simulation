```python
def HexToBin(hex):
    return [int(x) for x in bin(int(hex, 16))[2:].zfill(64)]  # bitarray can also be used here (list() not used- will become a list of ch instead of int)

def BinToHex(bin):
    return hex(int(''.join(map(str,bin)), 2))[2:].zfill(16).upper()

def Permute(block, perm_table):
    return [block[x-1] for x in perm_table] # list comprehension to apply permutation

def LeftShift(key, shift):
    return key[shift:]+key[:shift]

def KeySchedule(key):
    key_pc1 = Permute(key, PC1) # PC1
    K1, K2 = key_pc1[:28], key_pc1[28:]
    subkeys = []
    for shift in SHIFT:
        K1, K2 = LeftShift(K1, shift), LeftShift(K2, shift) # Left Shift
        subkey = Permute(K1+K2, PC2) # PC2
        subkeys.append(subkey)
    return subkeys

def XOR(block1, block2):
    return [x1^x2 for x1,x2 in zip(block1, block2)]

def SBox(block):
    plain = []
    for i in range(8):
        row= (block[i*6]<<1) + block[(i*6)+5]
        col= (block[(i*6)+1]<<3) + (block[(i*6)+2]<<2) + (block[(i*6)+3]<<1) + block[(i*6)+4]
        SBoxVal = S[i][row][col]
        plain += [int(x) for x in bin(SBoxVal)[2:].zfill(4)]
    return plain

def Feistel(plain, key):
    plain = Permute(plain, E) # Expansion
    plain = XOR(plain, key) # XOR with key
    plain = SBox(plain)     # SBox Substitution
    plain = Permute(plain, SP) # Straight Permutation

    return plain
```

```python
# DES 56
def encrypt56(plaintext, key):
    plainbin = HexToBin(plaintext)
    keybin = HexToBin(key)
    subkeys = KeySchedule(keybin)

    plainbin= Permute(plainbin, IP) # IP
    left, right = plainbin[:32], plainbin[32:]

    # 16 Feistel rounds
    for i in range(16):
        temp = left
        left = right
        right = XOR(temp, Feistel(right, subkeys[i]))
    
    cipherbin = Permute(right+left, FP)
    return BinToHex(cipherbin)

def decrypt56(ciphertext, key):
    cipherbin = HexToBin(ciphertext)
    keybin = HexToBin(key)
    subkeys = KeySchedule(keybin )
    cipherbin = Permute(cipherbin, IP) # IP
    left, right = cipherbin[:32], cipherbin[32:]
    
    for i in range(16):#16 rounds
        temp = XOR(left, Feistel(right, subkeys[16-(i+1)])) # apply in reverse
        left, right = right, temp

    plainbits = Permute(right + left, FP) # FP
    return BinToHex(plainbits)
```

```python
plaintext = "0123456789ABCDEF"  # input text must be 64 bits
key56 = "133457799bbcdff1" 
ciphertext56 = encrypt56(plaintext, key56)
decryptext56 = decrypt56(ciphertext56, key56) 
print("---------DES 56:---------- ")
print("Plain Text: "+plaintext)
print("Cipher Text: "+ciphertext56)
print("Decrypted Text: "+decryptext56)
print()
```

```python
# DES 112
def encrypt112(plaintext, key):
    spl = len(key)//2
    key1, key2 = key[:spl], key[spl:]
    cipher1 = encrypt56(plaintext, key1)
    cipher2 = decrypt56(cipher1, key2)
    return encrypt56(cipher2, key1)
def decrypt112(plaintext, key):
    spl = len(key)//2
    key1, key2 = key[:spl], key[spl:]
    plain1 = decrypt56(plaintext, key1)
    plain2 = encrypt56(plain1, key2)
    return decrypt56(plain2, key1)

#DES 168
def encrypt168(plaintext, key):
    spl = len(key)//3
    key1, key2, key3 = key[:spl], key[spl:2*spl], key[2*spl:]
    cipher1 = encrypt56(plaintext, key1)
    cipher2 = decrypt56(cipher1, key2)
    return encrypt56(cipher2, key3)
def decrypt168(plaintext, key):
    spl = len(key)//3
    key1, key2, key3 = key[:spl], key[spl:2*spl], key[2*spl:]
    plain1 = decrypt56(plaintext, key3)
    plain2 = encrypt56(plain1, key2)
    return decrypt56(plain2, key1)
```
